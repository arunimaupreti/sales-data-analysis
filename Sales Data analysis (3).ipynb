## Analyzing Gender

```python
data = pd.DataFrame({'Ratio':[len(df[df['Gender'] == 'M']), len(df[df['Gender'] == 'F'])]}, index = ['Male','Female'])
data
```

```python
data.plot.pie(y = 'Ratio', figsize = (6,6), autopct = '%.1f%%')
```

```python
df.groupby('Gender').size().plot(
    kind='bar', 
     
    title='Gender Ratio', 
    figsize=(6, 6)
)
```

```python
df.groupby('Gender')['Amount'].sum().plot(
    kind='pie', 
    autopct='%0.1f%%', 
    title='Average Purchase by Gender', 
    figsize=(6, 6)
)
```

```python
df.groupby('Gender')['Amount'].mean().plot(
    kind='pie', 
    autopct='%0.1f%%', 
    title='Average Purchase by Gender', 
    figsize=(6, 6)
)
```
